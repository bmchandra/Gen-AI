```python
import pandas as pd
import re
import string
import spacy
import emoji
```

#### Read csv

```python
data = pd.read_csv(r'sample.csv')
data.head()
```

#### lower case

```python
def to_lower(text):
    return text.lower()
```

```python
data['text'] = data['text'].apply(to_lower)
data[:3]
```

##### remove html tags

```python
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub('',text)
```

```python
data['text']=data['text'].apply(remove_html_tags)
data[:3]
```

### remove url

```python
def remove_url(text):
    pattern = re.compile(r"https://\S+|www\.S+")
    return pattern.sub('',text)
```

```python
data['text'] = data['text'].apply(remove_url)
data[:7]
```

##### string manipulation for removing punctuations ('!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~')

```python
def remove_punctuations(text):
    exclusion_list = string.punctuation
    for chars in exclusion_list:
        text = text.replace(chars,'')
    return text
```

```python
data['text'] = data['text'].apply(remove_punctuations)
```

#### expand Abbreviations [e.g. ASAP : As Soon as Possible]

```python
chat_words={
   " AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LMAO":"Laugh My A.. Off",
"FYI":"For your information"
}
chat_words
```

```python
def chat_conversion(text):
    new_txt=[]
    for w in text.split():
        if w.upper()in chat_words:
            new_txt.append(chat_words[w.upper()])
        else:
            new_txt.append(w)
    return ' '.join(new_txt)
```

```python
data['text'] = data['text'].apply(chat_conversion)
data
```

#### spelling Correction

```python
nlp = spacy.load("en_core_web_sm")
```

```python
from spellchecker import SpellChecker

# Create a SpellChecker instance
spell = SpellChecker()

# Define a function for spelling correction using SpaCy and pyspellchecker
def correct_spelling(text):
    # Tokenize the text using SpaCy
    doc = nlp(text)

    # Get the tokens and their corrected spellings
    corrected_tokens = [spell.correction(token.text) for token in doc]

    # Join the corrected tokens back into a string
    corrected_text = ' '.join(corrected_tokens)

    return corrected_text
```

##### remove stopwords using SPACY

```python
import spacy

# Load the SpaCy English model
nlp = spacy.load("en_core_web_sm")

# Define a function to remove stopwords
def remove_stopwords(text):
    # Tokenize the text using SpaCy
    doc = nlp(text)

    # Filter out tokens that are not stop words
    filtered_tokens = [token.text for token in doc if not token.is_stop]

    # Join the filtered tokens back into a string
    result_text = ' '.join(filtered_tokens)

    return result_text

# Example usage
text_with_stopwords = "This is an example sentence with some stopwords."
text_without_stopwords = remove_stopwords(text_with_stopwords)

print("Original Text:", text_with_stopwords)
print("Text without Stopwords:", text_without_stopwords)
```

```python
data['text']=data['text'].apply(remove_stopwords)
data.head()
```

#### remove EMOJIs

```python
def remove_emoji(text):
    clean_text=emoji.demojize(text)
    return clean_text
```

```python
data['text']=data['text'].apply(remove_emoji)
data.head()
```

#### Lemmatization using SPACY 

```python
def lemmatize_text(text):
    # Process the text using SpaCy
    doc = nlp(text)
    
    # Extract the lemmatized forms of each token
    lemmatized_tokens = [token.lemma_ for token in doc]
    
    # Join the lemmatized tokens back into a string
    result_text = ' '.join(lemmatized_tokens)
    
    return result_text
```

```python
data['text']=data['text'].apply(lemmatize_text)
data.head()
```
